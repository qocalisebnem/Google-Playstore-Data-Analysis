```python
import numpy as np
def size_mb(x):
    
    if isinstance(x,str):
        if ',' in x and x.endswith('M'):
                a = x.replace(',','') 
                return float(a.replace('M',''))
        elif ',' in x and x.endswith('k'):
            b=x.replace(',','')
            return float(b.replace('k',''))/1024
        elif x.endswith('M'):
            return float(x.replace('M',''))
        elif x.endswith('k'):
            return float(x.replace('k',''))/1024
        else:
            return np.nan
    else:
        return np.nan
df['Size']=df['Size'].apply(size_mb)
print(df['Size'].unique().tolist())
```

```python
import pandas as pd

def detect_outliers_iqr(df, column):
    """
    Verilən sütun üzrə IQR metodu ilə outlier-ları tapır.
    :param df: pandas DataFrame
    :param column: yoxlanılacaq sütun adı
    :return: outlier-ların siyahısı
    """
    Q1 = df[column].quantile(0.25)  # 25%-lik kvartil
    Q3 = df[column].quantile(0.75)  # 75%-lik kvartil
    IQR = Q3 - Q1                   # IQR hesablanır

    # Aşağı və yuxarı limitləri təyin edirik
    lower_limit = Q1 - 1.5 * IQR
    upper_limit = Q3 + 1.5 * IQR

    # Outlier-ları seçirik
    outliers = df[(df[column] < lower_limit) | (df[column] > upper_limit)]

    return outliers[column].tolist()
    
adr_outliers = detect_outliers_iqr(df, "Maximum Installs")
print(f"Maximum Installs üzrə outlier sayı: {len(adr_outliers)}")
print("Nümunə outlier-lar:", adr_outliers[:20])
```

```python
# Kolmogorov-Smirnov Testi
from scipy.stats import kstest
from scipy.stats import zscore

max_installs_test=df['Maximum Installs']
max_installs_zscore=zscore(max_installs_test)

ks_stat,ks_p=kstest(max_installs_zscore,'norm')
print(f'p_value:{ks_p}')
if ks_p>0.05:
    print('Maximum Installs sutunu normal paylanib')
else:
    print('Maximum Installs sutunu normal paylanmayib')
```

```python
from scipy.stats import kstest
from scipy.stats import zscore

max_installs_test=df['Price']
max_installs_zscore=zscore(max_installs_test)

ks_stat,ks_p=kstest(max_installs_zscore,'norm')
print(f'p_value:{ks_p}')
if ks_p>0.05:
    print('Price sutunu normal paylanib')
else:
    print('Price sutunu normal paylanmayib')
```
